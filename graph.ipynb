```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_excel('k.xlsx')

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize = (7,7))
ax.scatter(data['theta'][data['treatment'] == 'IHC'], data['r'][data['treatment'] == 'IHC'], s=40, c = 'orange', edgecolors = 'black', linewidth = 0.5, label ='IHC')
ax.scatter(data['theta'][data['treatment'] == 'OHC1'], data['r'][data['treatment'] == 'OHC1'], s=40, c = 'orange', edgecolors = 'black', linewidth = 0.5, label ='OHC1')
ax.scatter(data['theta'][data['treatment'] == 'OHC2'], data['r'][data['treatment'] == 'OHC2'], s=40, c = 'blue', edgecolors = 'black', linewidth = 0.5, label = 'OHC2')
ax.scatter(data['theta'][data['treatment'] == 'OHC3'], data['r'][data['treatment'] == 'OHC3'], s=40, c = 'green', edgecolors = 'black', linewidth = 0.5, label = 'OHC3')

ax.set_rmax(0.5)

plt.xlabel('OHC', fontsize=20)
#ax.legend(loc='upper right', fontsize=30, bbox_to_anchor = (1.1,1.1))


#plt.savefig('OHC_apex_L.svg', dpi=1200)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd



data = pd.read_excel('k.xlsx')
fig, ax = plt.subplots()
p = sns.scatterplot(data=data, x='x', y='y', ax=ax)

ax.set_ylim(-0.5,0.5)
ax.set_xlim(-0.5,0.5)
sns. set_style("darkgrid")

plt.axvline(0, c='black', ls='--')
plt.axhline(0, c='black', ls='--')

plt.xlabel('X')
plt.ylabel('Y')
plt.grid()

sns.set(rc={'figure.figsize':(5,5)})
```

```python
import numpy as np
import pandas as pd
import seaborn as sns

data = pd.read_excel('k.xlsx')

sns.set_theme()

g = sns.FacetGrid(data, hue="treatment",
                  subplot_kws=dict(projection='polar'), height=4.5,
                  sharex=False, sharey=False, despine=False)


# Draw a scatterplot onto each axes in the grid
g.map(sns.scatterplot, "theta", "r", legend='full')
```
