# KINOCILIA POSITION SCRIPT
This script is to plot a polar plot for kinocilia. You can use this same script to analyze files at different locations by copy pasting the path of the files location. Save the result csv files from the same image in same folder and use it as working directory. There are 5 parts to this script and all the cells work independtly of each other execpt 1 and 2
1) Plots only IHC kinocilia positions as a polar plot
2) Plots only OHC1 kinocilia positions as a polar plot
3) Plots only OHC2 kinocilia positions as a polar plot
4) Plots only OHC3 kinocilia positions as a polar plot
5) Plots combined kinocilia positions as a polar plot

Run following two cells first

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os
import seaborn as sns
```

## Plot for IHC
make sure you save the csv result file as IHC.csv

```python
#load the raw measurement csv file for IHC
data = pd.read_csv('IHC.csv')

#restructure the dataframe
data.drop(columns=['Ch','Slice'], inplace=True)
data.columns = ['index', 'X','Y']

#create a dataframe for X and Y pos values of kinocilia
length = int(len(data)/5)
cells = list(range(0,length,1))

df = []

for i in range(length):
    x1 = float(data.iloc[5*i]['X'])
    x2 = float(data.iloc[5*i+1]['X'])
    x3 = float(data.iloc[5*i+2]['X'])
    y1 = float(data.iloc[5*i]['Y'])
    y4 = float(data.iloc[5*i+3]['Y'])
    y5 = float(data.iloc[5*i+4]['Y'])

    X = (x1-x2)/(x3-x2) - 0.5
    Y = (y1-y4)/(y5-y4) - 0.5

    df.append([X,Y,'IHC'])
kinocilia = pd.DataFrame(df, columns = ['X','Y','HC'])

#calculate r and theta for polar plot and add them to the dataframe
kinocilia['r'] = np.hypot(kinocilia['X'], kinocilia['Y'])  # radius
kinocilia['theta'] = np.arctan2(kinocilia['Y'],kinocilia['X'])  # angle in radians

#plot a polar xy scatter plot for kinocilia position
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize = (7,7))
ax.scatter(kinocilia['theta'],kinocilia['r'],s=40, c = 'orange', edgecolors = 'black', linewidth = 0.5, label ='IHC')
ax.set_rmax(0.5)

plt.xlabel('IHC', fontsize=20)
plt.savefig('IHC.svg', dpi=1200)
```

## Plot for combined IHC and OHCs
No need to repeat the imagej analysis for ihc and ohc again. This script uses the IHC.csv and OHC1/2/3.csv files to create a combined polar plot

```python
#load the raw measurement csv file for IHCs and OHCs
IHC = pd.read_csv('IHC.csv')
OHC1 = pd.read_csv('OHC1.csv')
OHC2 = pd.read_csv('OHC2.csv')
OHC3 = pd.read_csv('OHC3.csv')

IHC['HC'] = 'IHC'
OHC1['HC'] = 'OHC1'
OHC2['HC'] = 'OHC2'
OHC3['HC'] = 'OHC3'

data = pd.concat([IHC, OHC1, OHC2, OHC3], axis=0)
data.reset_index(drop=True, inplace=True)

#restructure the dataframe
data.drop(columns=['Ch','Slice'], inplace=True)
data.columns = ['index', 'X','Y','HC']

#create a dataframe for X and Y pos values of kinocilia
length = int(len(data)/5)
df = []

for i in range(length):
    x1 = float(data.iloc[5*i]['X'])
    x2 = float(data.iloc[5*i+1]['X'])
    x3 = float(data.iloc[5*i+2]['X'])
    y1 = float(data.iloc[5*i]['Y'])
    y4 = float(data.iloc[5*i+3]['Y'])
    y5 = float(data.iloc[5*i+4]['Y'])

    X = (x1-x2)/(x3-x2) - 0.5
    Y = (y1-y4)/(y5-y4) - 0.5

    df.append([X,Y,data.iloc[5*i]['HC']])

kinocilia = pd.DataFrame(df, columns = ['X','Y','HC'])

#calculate r and theta for polar plot and add them to the dataframe
kinocilia['r'] = np.hypot(kinocilia['X'], kinocilia['Y'])  # radius
kinocilia['theta'] = np.arctan2(kinocilia['Y'],kinocilia['X'])  # angle in radians

#save the transformed values in a csv file
kinocilia.to_csv('kinocilia.csv')

#plot a polar xy scatter plot for kinocilia position
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize = (7,7))
ax.scatter(kinocilia['theta'][kinocilia['HC'] == 'IHC'], kinocilia['r'][kinocilia['HC'] == 'IHC'], s=40, c = 'orange', edgecolors = 'black', linewidth = 0.5, label ='IHC')
ax.scatter(kinocilia['theta'][kinocilia['HC'] == 'OHC1'], kinocilia['r'][kinocilia['HC'] == 'OHC1'], s=40, c = 'orange', edgecolors = 'black', linewidth = 0.5, label ='OHC1')
ax.scatter(kinocilia['theta'][kinocilia['HC'] == 'OHC2'], kinocilia['r'][kinocilia['HC'] == 'OHC2'], s=40, c = 'blue', edgecolors = 'black', linewidth = 0.5, label = 'OHC2')
ax.scatter(kinocilia['theta'][kinocilia['HC'] == 'OHC3'], kinocilia['r'][kinocilia['HC'] == 'OHC3'], s=40, c = 'green', edgecolors = 'black', linewidth = 0.5, label = 'OHC3')

ax.set_rmax(0.5)

ax.legend(loc='upper right', fontsize=10, bbox_to_anchor = (1.1,1.1))
plt.xlabel('combined', fontsize=20)
plt.savefig('combined.svg', dpi=1200)
```

KDE plot to show a rough distribution of kinocilia using heat map

```python
kinocilia = pd.read_csv('kinocilia.csv')

i = kinocilia[kinocilia['HC']=='IHC']
o1 = kinocilia[kinocilia['HC']=='OHC1']
o2 = kinocilia[kinocilia['HC']=='OHC2']
o3 = kinocilia[kinocilia['HC']=='OHC3']

fig, ax = plt.subplots(2,2,subplot_kw={'projection':'polar'},figsize=(14,14))

ax[0, 0].set_title('IHC')
ax[0, 1].set_title('OHC1')
ax[1, 0].set_title('OHC2')
ax[1, 1].set_title('OHC3')

sns.kdeplot(ax = ax[0,0], data=i, x='theta',y='r',fill=True,cbar=True, color = 'blue')
sns.kdeplot(ax = ax[0,1], data=o1, x='theta',y='r',fill=True,cbar=True, color = 'green')
sns.kdeplot(ax = ax[1,0], data=o2, x='theta',y='r',fill=True,cbar=True, color = 'red')
sns.kdeplot(ax = ax[1,1], data=o3, x='theta',y='r',fill=True,cbar=True, color = 'cyan')

for ax in ax.flat:
    ax.set(xlabel='', ylabel='')
    
plt.savefig('kdeplot.svg', dpi=1200)
```
